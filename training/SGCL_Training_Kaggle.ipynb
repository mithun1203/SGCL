## ğŸ“Š Step 5: View Results

````python
import json

# Load training summary
with open('sgcl_checkpoints/training_summary.json', 'r') as f:
    summary = json.load(f)

print("\nğŸ“ˆ SG-CL Training Summary")
print("=" * 50)
print(f"Total tasks: {summary['num_tasks']}")
print(f"Total steps: {summary['total_steps']}")
print(f"\nPer-task metrics:")

for i, metrics in enumerate(summary['task_metrics']):
    avg_loss = sum(metrics['loss']) / len(metrics['loss'])
    print(f"\n  Task {i}:")
    print(f"    Samples: {metrics['samples']}")
    print(f"    Conflicts detected: {metrics['conflicts']}")
    print(f"    Guard-rails generated: {metrics['guardrails']}")
    print(f"    Avg loss: {avg_loss:.4f}")
````
